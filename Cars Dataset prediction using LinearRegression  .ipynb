```python
import pandas as pd 
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns
print('imported')
```

```python
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv'
df=pd.read_csv(path)
df.head()
```

```python
headers=['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','heighth','curb-weight','engine-type','num-of-cylendres','engine-size','fuel-sys','bore','strock','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
```

```python
df.columns=headers
df.head(10)
```

```python
#convert "?" to NaN 
df.replace('?',np.nan,inplace=True)
df.head(10)
```

```python
miss_data=df.isnull()
df.head()
```

```python
# False stands for not missing value
# True stands for missing value 
for column in miss_data.columns.values.tolist():
    print(column)
    print(miss_data[column].value_counts())
    print()
```

```python
avg_hp=df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.NaN,avg_hp,inplace=True)
```

```python
avg_pr=df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.NaN,avg_pr,inplace=True)
```
